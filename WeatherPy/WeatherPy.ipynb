```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
from scipy import stats
import gmaps
import csv
import os
import random
from citipy import citipy


# Import API key
from api_keys import weather_api_key
```

## Generate Cities List

```python
cities=[]
lat_long=[]
lat=np.random.uniform(-90,90,1500)
long=np.random.uniform(-180,180,1500)
lat_long=zip(lat,long)

for lat_long in lat_long:
    city = citipy.nearest_city(lat_long[0], lat_long[1]).city_name
    if city not in cities:
        cities.append(city)
cities

# calculate length cities
len(cities)
        
```

```python
import json

# Save config information.
url = "http://api.openweathermap.org/data/2.5/weather?"
units="imperial"

# Build partial query URL
query_url = url+"appid="+weather_api_key+"&units="+units+"&q="
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
import json

Lat=[]
Lng=[]
City=[]
Cloudiness=[]
Country=[]
Date=[]
Humidity=[]
Max_Temp=[]
Wind_Speed=[]

counter=0
print("Beginning Data Retrieval")
print("-------------------------")

for city in cities:
    response = requests.get(query_url + city).json()
    if counter < 5:
        counter+=1
    else:
        counter=0 
    print("Processing Record {} | {}".format(counter,city))
    try:
        Lat.append(response['coord']['lat'])
        Max_Temp.append(response['main']['temp_max'])
        Lng.append(response['coord']['lon'])
        City.append(city)
        Wind_Speed.append(response['wind']['speed'])
        Humidity.append(response['main']['humidity'])
        Date.append(response['dt'])
        Country.append(response['sys']['country'])
        Cloudiness.append(response['clouds']['all'])
        
    except KeyError:
        print("City not found. Skipping...")
        pass
```

### Convert Raw Data to DataFrame
* Export the city data into a .csv.
* Display the DataFrame

```python
city_weather_dict={
    "City":City,
    "Lat":Lat,
    "Lng":Lng,
    "Max_Temp": Max_Temp ,
    "Humidity":Humidity,
    "Cloudiness":Cloudiness,
    "Wind_Speed":Wind_Speed,
    "Country":Country,
    "Date":Date
}

city_weather_df=pd.DataFrame.from_dict(city_weather_dict,orient='index')
city_weather_df=city_weather_df.transpose()
city_weather_df
```

```python
city_weather_df[["Lat","Lng","Max_Temp","Humidity","Cloudiness", "Wind_Speed"]]=city_weather_df[["Lat","Lng","Max_Temp","Humidity","Cloudiness", "Wind_Speed"]].astype("int64")
city_weather_df.dtypes
```

## Linear Regression

```python
from scipy.stats import linregress
```

```python
northern=city_weather_df.loc[city_weather_df['Lat'] >=0 ]
southern=city_weather_df.loc[city_weather_df['Lat'] <=0.1]
southern.head()
```

####  Northern Hemisphere - Max Temp vs. Latitude Linear Regression

```python
x_values=northern["Lat"]
y_values=northern["Max_Temp"]
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values= x_values*slope+intercept
print(f"The r-squared is:{rvalue**2}")
line_eq="y="+str(round(slope,2))+"X +"+ str(round(intercept,2))
plt.scatter(x_values,y_values)
plt.plot(x_values, regress_values,"r-")
plt.annotate(line_eq,(6,10),fontsize=15,color="red")
plt.xlabel("Latitude")
plt.ylabel("Max_Temp")
plt.show()
```

####  Southern Hemisphere - Max Temp vs. Latitude Linear Regression
